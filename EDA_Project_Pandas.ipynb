# **Exploratory Data Analysis- World Population**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Set visualization style
sns.set_style("whitegrid")
```

```python
### 1️⃣ Load Data ###
df = pd.read_csv('world_population.csv')
```

```python
### 2️⃣ Data Overview ###
print("\nBasic Data Overview:")
df.info()

```

```python
print("\nFirst 5 rows of the dataset:")
df.head()
```

```python
print("\nSummary Statistics:")
df.describe()
```

```python
# Check for missing values
print("\nMissing Values:")
df.isnull().sum()
```

```python
# Check unique values in 'Continent' column
print("\nUnique Continents:", df['Continent'].unique())
```

```python
# Ensure numeric columns are in correct format
df['2022 Population'] = pd.to_numeric(df['2022 Population'], errors='coerce')
```

```python
### 3️⃣ Descriptive Statistics & Insights ###
# Calculate Population Growth Rate
df['Growth Rate (%)'] = ((df['2022 Population'] - df['2000 Population']) / df['2000 Population']) * 100
```

```python
# Calculate Population Density
df['Density'] = df['2022 Population'] / df['Area (km²)']
```

```python
# Top 10 most populated continents in 2022
top_continents = df.sort_values(by="World Population Percentage", ascending=False).head(10)
print("\nTop 10 Most Populated Continents (2022):")
top_continents[['Continent', '2022 Population', 'World Population Percentage']]
```

```python
# Compute correlation matrix (for numeric columns only)
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

```python
### 4️⃣ Data Visualizations ###
# Heatmap of Correlations
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Population Data")
plt.show()
```

```python
# Population Growth Trend per Continent
df_grouped = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']].mean()
df_transposed = df_grouped.transpose()
```

```python
plt.figure(figsize=(12, 6))
for column in df_transposed.columns:
    plt.plot(df_transposed.index, df_transposed[column], label=column)
plt.legend()
plt.title("Average Population Growth Per Continent (1970-2022)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
# Boxplot for Population Distribution
plt.figure(figsize=(12, 6))
df[['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']].boxplot()
plt.title("Boxplot of Population Distributions (1970-2022)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df['Continent'], y=df['Growth Rate (%)'], hue=df['Continent'], palette='viridis', legend=False)
plt.title("Population Growth Rate per Continent (2000-2022)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot: Population vs Area
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Area (km²)'], y=df['2022 Population'], hue=df['Continent'])
plt.xscale('log')
plt.yscale('log')
plt.title("Population vs Land Area")
plt.show()
```

```python
### 5️⃣ Conclusion & Next Steps ###
print("\nKey Takeaways:")
print("- Asia consistently holds the highest population, while Oceania remains the least populated.")
print("- The population has increased significantly across all continents, especially in Africa and Asia.")
print("- The correlation heatmap shows strong positive relationships between historical and recent population values.")
```
